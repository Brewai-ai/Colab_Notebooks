```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Step 1: Loading the dataset
data = pd.read_csv('Battery_RUL.csv')
```

```python
# Step 2: Splitting the data into features (X) and target (y)
X = data.drop(columns=['RUL'])  # Features excluding the RUL column
y = data['RUL']  # Target variable (Remaining Useful Cycles)
```

```python

# Step 3: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Step 4: Selecting Linear regression and training model on it
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
# Step 5: Making predictions on the test set
y_pred = model.predict(X_test)

```

```python
# Step 6: finding the error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

```

```python
plt.figure(figsize=(40, 10))
plt.scatter(y_test, y_test, color='blue', label='Actual RUL')
plt.scatter(y_test, y_pred, color='red', label='Predicted RUL')
plt.xlabel('Sample Index')
plt.ylabel('Remaining Useful Cycles (RUL)')
plt.title('Actual RUL vs. Predicted RUL')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(r2_score(y_test, y_pred))
```
